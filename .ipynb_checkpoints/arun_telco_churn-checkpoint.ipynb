# We need to bin tenure and plot month to month, tenure, churn. Maybe those with high tenure and month to month are churning a lot.

```r
df %>% 
    mutate(tenure = case_when(tenure <= quantile(tenure, 0.33) ~ 'New customer',
                              tenure <= quantile(tenure, 0.66) ~ 'Medium',
                              TRUE ~ 'Old customer')) %>% 
    ggplot(aes(x = Churn)) +
    geom_bar(aes(fill = Churn)) +
    facet_wrap(tenure ~ Contract, scales = 'free_y')
```

```r
df %>% 
    mutate(MonthlyCharges = case_when(MonthlyCharges <= quantile(MonthlyCharges, 0.33) ~ 'Low Charges',
                              MonthlyCharges <= quantile(MonthlyCharges, 0.66) ~ 'Medium Charges',
                              TRUE ~ 'High charges')) %>% 
    ggplot(aes(x = Churn)) +
    geom_bar(aes(fill = Churn)) +
    facet_wrap(MonthlyCharges ~ Contract, scales = 'free_y')
```

```r
split_obj <- rsample::initial_split(df, prop = 0.8, strata = 'Churn')
```

```r
train = rsample::training(split_obj)
test = rsample::testing(split_obj)
```

```r
df %>% 
    map(levels)
```

```r
recipe_obj <- df %>% 
    recipe(Churn ~ ., data = train) %>%
    step_knnimpute(TotalCharges) %>% 
    step_YeoJohnson(all_numeric()) %>%
    step_center(all_numeric()) %>% 
    step_scale(all_numeric()) %>%
    step_mutate(SeniorCitizen = SeniorCitizen %>% as.numeric()) %>% 
    step_dummy(all_nominal(), one_hot = T) %>% 
    step_mutate(Churn_Yes = Churn_Yes %>% as_factor())

train_processed <- recipe_obj %>%
    prep() %>% 
    bake(train)

test_processed <- recipe_obj %>%
    prep() %>% 
    bake(test)

    
```

```r
col_names <- train_processed %>% colnames()
col_names %>% enframe
```

```r
drop_cols <- col_names[c(5,8,9,11,13,14,16,20,21,22,24,26,27,29,30,31,33,34,36,38,41,43,47)]
drop_cols
```

```r
train_processed <- train_processed %>% select(-drop_cols)
test_processed <- test_processed %>% select(-drop_cols)
```

#### We can get monthly charges by looking at the services which the customer uses.
#### We can get tenure from monthly charges and total charges.
#### So lets drop monthly charges and total charges.

```r
drop_cols <- append(drop_cols, c('MonthlyCharges', 'TotalCharges'))

# Lets modify the recipe
recipe_obj <- recipe_obj %>% 
                step_rm(drop_cols) 

train_processed <- recipe_obj %>%
    prep() %>% 
    bake(train)

test_processed <- recipe_obj %>%
    prep() %>% 
    bake(test)
```

#### Our baseline accuracy is 0.64.

```r
metric_set_class <- metric_set(accuracy ,bal_accuracy, precision, recall, f_meas)
metric_set_prob <- metric_set(roc_auc, mn_log_loss)

calc_metrics <- function(model, data = test_processed){
    
    class_metrics <- model %>% 
    predict(data) %>%  
    bind_cols(data %>% select(Churn_Yes))  %>%
    metric_set_class(truth = Churn_Yes, estimate = .pred_class)
    
    prob_metrics <- model %>% 
    predict(data, type = 'prob') %>%  
    bind_cols(data %>% select(Churn_Yes))  %>%
    metric_set_prob(truth = Churn_Yes, .pred_1)
    
    return(bind_rows(class_metrics, prob_metrics))
    
}
```

```r
folds <- rsample::vfold_cv(data = train_processed, v=5, strata = 'Churn_Yes')

rec_form <-
  tune_grid(
    recipe_obj,
    model = svm_mod,
    resamples = folds,
    metrics = metric_set_prob
  )
```

```r
model_1 <- logistic_reg(mode ='classification', penalty = 100, mixture = 0.25) %>% 
    set_engine('glmnet', standardize = FALSE) %>% 
    fit(Churn_Yes ~ ., data = train_processed)
```

```r
calc_metrics(model_1)
```

#### Even with a simple model like logistic regression we are getting good results.

```r
coefs <- tidy(model_1) %>% 
            filter(p.value <= 0.05) %>% 
            slice(2:nrow(.))
coefs
```

```r
coefs %>%
    mutate(term = term %>% as_factor() %>% fct_reorder(estimate)) %>% 
    ggplot(aes(y = term, x = estimate)) +
    geom_vline(xintercept = 0) +
    geom_segment(aes(xend = 0, yend = term), color = ifelse(coefs$estimate >= 0 , 'green', 'red')) +
    geom_label(aes(label = round(estimate,2))) +
    theme_tq()+
    labs(title = 'Coefficients obtained from logistic regression', y = '')
```

#### The results for logistic regression are similar to the results in correlation analysis.

```r
model_2 <- rand_forest(mode ='classification') %>% 
    set_engine('randomForest', importance = TRUE) %>% 
    fit(Churn_Yes ~ ., data = train_processed)
```

```r
randomForest::importance(model_2$fit)[,'MeanDecreaseGini'] %>% 
    enframe() %>%
    mutate(name = name %>% fct_reorder(value)) %>% 
    slice(1:10) %>% 
    ggplot(aes(x = name, y = value)) +
    geom_col() +
    geom_label(aes(label = round(value,2))) +
    theme_tq() +
    coord_flip()
```

```r
calc_metrics(model_2)
```

```r
model_3 <- boost_tree(mode ='classification') %>% 
    set_engine('xgboost', importance = TRUE) %>% 
    fit(Churn_Yes ~ ., data = train_processed)
```

```r
calc_metrics(model_3)
```
